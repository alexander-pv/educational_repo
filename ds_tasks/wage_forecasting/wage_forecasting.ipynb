### Прогнозирование уровня средней заработной платы в России

В файле записаны данные о заработной плате за каждый месяц с января 1993 по август 2016.

Необходимо проанализировать данные, подобрать для них оптимальную прогнозирующую модель в классе ARIMA и построить прогноз на каждый месяц на два года вперёд от конца данных.

Действуем по следующим шагам:

   1. Визуальный анализ ряда
   2. Стабилизация дисперсии (при необходимости)
   3. Выбор порядка дифференцирования
   4. Выбор начальных приближений для p,q,P,Q
   5. Обучение и сравнение моделей-кандидатов, выбор победителя
   6. Анализ остатков построенной модели, при необходимости — её модификация: если остатки получаются смещёнными, прогноз нужно скорректировать на константу; если остатки нестационарны или автокоррелированы, можно попробовать расширить область, в которой подбираются значения p,q,P,Q
   7. Прогнозирование с помощью финальной модели.

````python
import os
import tqdm
import warnings
import datetime

import pandas as pd
import numpy as np
import scipy as sp
import statsmodels.api as sm

from scipy import stats

import pylab
import matplotlib.pyplot as plt
import seaborn as sns

from itertools import product

%config InlineBackend.figure_format = 'retina'
sns.set()
warnings.filterwarnings('ignore')
````

````python
df = pd.read_csv('WAG_C_M.csv', sep=';')
df['month'] = pd.to_datetime(df['month'], dayfirst=True)
df.head()
````

````python
def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
````

1. Визуальный анализ ряда

````python
fig = plt.figure(figsize=(10,6))
plt.plot(df['month'], df['WAG_C_M'])
plt.title('Динамика уровня средней заработной платы в России')
plt.show()
````

В динамике заработной платы присутствуют растущая трендовая компонента, а также годовая сезонность.
При этом видно, что колебания внутри года становятся существеннее со временем, т.е. идет рост дисперсии заработной платы. Соответственно, свойства временного ряда меняются со временем (в первую очередь, матожидание и дисперсия), ряд не стационарен.

Формально можем проверить с помощью теста Дики-Фуллера:
```
H0: ряд нестационарен;
H1: ряд стационарен;
```

````python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['WAG_C_M'])[1])
````

p-value>0.05, соответственно, гипотеза H0 не отвергается, ряд нестационарен.

Попробуем разбить исходный ряд на сезонную компоненту, тренд и ошибку с помощью  STL-декомпозиции, период по 12 месяцев в году.

````python
def make_stl_decompose(df, column, period, start_index=0):
    plt.figure(figsize=(15, 9))
    pylab.rcParams['figure.figsize'] = (15, 9)
    df.sort_index(inplace=True)
    sm.tsa.seasonal_decompose(x=df[column].iloc[start_index:], period=period).plot()
    plt.tight_layout()
    plt.show()
````

````python
make_stl_decompose(df, 'WAG_C_M', 12)
````

Отчетливо выделяется тренд, ошибка находится в районе нуля, однако есть ее разнонаправленное смещение в начале и в конце ряда. Поскольку исходный ряд нестационарен, попробуем для начала провести сезонное дифференцирование, выполнить STL-декомпозицию ряда, а также протестировать дифференцированный ряд на нестационарность.

````python
# Проведем сезонное дифференцирование
days_shift=12
df['wag_c_m_diff'] = df['WAG_C_M']-df['WAG_C_M'].shift(days_shift)
````

````python
make_stl_decompose(df, 'wag_c_m_diff', 12, 12)
````

````python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['wag_c_m_diff'].iloc[12:])[1])
````

Ряд с сезонным дифференцированием также нестационарен, в STL-декомпозиции дисперсия ошибки снизилась и стала менее систематической в конце ряда, в свою очередь тренд стал более детальным после нивелирования годовой сезонности. 

Применим обычное дифференцирование, чтобы попытаться сделать ряд стационарным, и снова рассмотрим результат STL-декомпозиции и теста Дики-Фуллера.

````python
df['wag_c_m_diff2'] = df['wag_c_m_diff']-df['wag_c_m_diff'].shift(1)
````

````python
make_stl_decompose(df, 'wag_c_m_diff2', 12, 13)
````

````python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['wag_c_m_diff2'].iloc[13:])[1])
````

Мы получили стационарный ряд `wag_c_m_diff2`, согласно тесту Дики-Фуллера, однако на графике видно, что дисперсия этого ряда отличается в разные моменты времени: первая половина ряда менее волатильна, чем его вторая половина. Это также отражается на ошибке и в несколько большой яме тренда, опускающемся в отрицательную область.

Попробуем исправить эту ситуацию, предварительно осуществив преобразование Бокса-Кокса для стабилизации дисперсии ряда.

2. Стабилизация дисперсии

````python
df['wag_c_m_box'], lmbda = stats.boxcox(df['WAG_C_M'])
print(f'Lambda for Box-Cox: {lmbda}')
````

````python
fig = plt.figure(figsize=(10,6))
plt.plot(df['month'], df['wag_c_m_box'])
plt.title('Динамика уровня средней заработной платы в России после преобразвоания Бокса-Кокса')
plt.show()
````

````python
df['wag_c_m_box_diff'] = df['wag_c_m_box']-df['wag_c_m_box'].shift(12)
````

````python
make_stl_decompose(df, 'wag_c_m_box_diff', 12, 12)
````

````python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['wag_c_m_box_diff'].iloc[13:])[1])
````

После сезонного дифференцирования видим, что тренд стал колебаться намного меньше, однако ошибка все еще содердит в себе некоторую структуру несмотря на то, что мы можем отвергать нулевую гипотезу о нестационарности ряда. 

Попробуем добавить обычное дифференцирование.

````python
df['wag_c_m_box_diff2'] = df['wag_c_m_box_diff']-df['wag_c_m_box_diff'].shift(1)
````

````python
make_stl_decompose(df, 'wag_c_m_box_diff2', 12, 13)
````

````python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['wag_c_m_box_diff2'].iloc[13:])[1])
````

Последовательность преобразований ряда `Стабилизация дисперсии->Сезонное дифференцирование->Обычное дифференцирование` привели к стационарному ряду, в котором тренд в STL-декомпозиции колеблется близко к нулю, а в ошибке нельзя увидеть некоторую нераскрытую структуру ряда. 

3. Выбор порядка дифференцирования

Рассмотрим графики автокорреляционной и частичной автокорреляционной функций. (ACF и PACF)

````python
plt.figure(figsize=(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df['wag_c_m_box_diff2'][13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df['wag_c_m_box_diff2'][13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
````

4. Выбор начальных приближений для p,q,P,Q

Зададим начальные приближения для ARIMA-модели:

Начальные приближения: Q=0, q=5, P=0, p=5

````python
ps = range(0, 6)
d=1
qs = range(0, 6)

Ps = range(0, 1)
D=1
Qs = range(0, 1)
````

````python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
````

5. Обучение и сравнение моделей-кандидатов, выбор победителя

````python
def find_best_sarima(parameters_list, y_series):
    """
    Подбор в цикле параметров модели SARIMA по данным начальных приближений.
    """
    results = []
    best_aic = float("inf")
    
    for param in tqdm.tqdm(parameters_list):
        #try except нужен, потому что на некоторых наборах параметров модель не обучается
        try:
            model=sm.tsa.statespace.SARIMAX(y_series, order=(param[0], d, param[1]), 
                                            seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
        #Выводим параметры, на которых модель не обучается и переходим к следующему набору
        except ValueError:
            print('Wrong parameters:', param)
            continue
        aic = model.aic
        #Сохраняем лучшую модель, aic, параметры
        if aic < best_aic:
            best_model = model
            best_aic = aic
            best_param = param
        results.append([param, model.aic])
    return results, best_param, best_model
````

````python
results, best_param, best_model = find_best_sarima(parameters_list, df['wag_c_m_box'])
print(best_param)
````

````python
# Статистика по лучшей модели
best_model.summary()
````

6. Проанализируем остатки модели:

````python
plt.figure(figsize=(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel('Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
````

Критерий Стьюдента не отвергает нулевую гипотезу с alpha=0.05 о том, что среднее значение остатков равно нулю, т.е. остатки модели несмещенные.

Остатки не автокоррелированы, что можно видеть по коррелелограмме и тесту Льюнга-Бокса (нулевая гипотеза об отсутствии автокорреляции не отвергается)

Кроме того, остатки представляют собой стационарный ряд, что можно видеть как по графику остатков визуально, так и по достигаемому кровню значимости теста Дики-Фуллера.

7. Прогнозирование заработной платы итоговой моделью

Используя новые данные по заработной плате http://sophist.hse.ru/hse/1/tables/WAG_M.htm,
проведем прогноз на 12 месяцев вперед и оценим его качество по фактическим данным.

````python
new_data = {'month':[str(x.date()) for x in pd.date_range(start='2016-09-01', 
                                                          end='2017-08-01', 
                                                          freq='MS')],
            
            'WAG_C_M':[35843.0, 35749.0, 36195.0, 47554.0,
                       34422.0, 35497.0, 37899.0, 39225.0, 
                       39679.0, 41454.0, 38073.0, 37099.0 ]}
````

````python
df_new = pd.DataFrame(new_data)
df_new['month'] = pd.to_datetime(df_new['month'], dayfirst=True)
df_new.reset_index(inplace=True, drop=True)
````

````python
forecast_horizont = '2017-08-01'

# Зададим размах прогнозирования с date_range, MS-частота списка дат: начало месяца
date_list = pd.date_range(start=str(df['month'].max().date()), 
                          end=forecast_horizont,
                          freq='MS'
                         )[1:]

date_list
````

````python
#  Формируем датасет с горизонтом прогнозирования
history_length = df.shape[0]

df_future = pd.DataFrame(index=date_list, columns=df.columns)
df = pd.concat([df, df_future])
df.reset_index(inplace=True, drop=True)
````

````python
# Осуществим прогноз
df['forecast'] = invboxcox(best_model.predict(start=history_length, 
                                              end=history_length+len(date_list)), 
                                lmbda
                               )

# Добавим данные месяцев в датасет
df.loc[df.index.isin(list(range(history_length, history_length+len(date_list)))),
       'month'] = date_list

# Добавим реальные данные в датасет
df.loc[df.index.isin(list(range(history_length, history_length+len(date_list)))),
       'WAG_C_M'] = df_new['WAG_C_M'].values
````

````python
df.tail(10)
````

````python
#  Нарисуем прогноз и факт на графике
plt.figure(figsize=(15,7))
df['WAG_C_M'].iloc[250:].plot(label='Real data')
df['forecast'].iloc[250:].plot(color='r', label='Forecast')
plt.ylabel('Wage by month')
plt.legend()
plt.show()
````

Для оценки прогноза всопользуемся метрикой MAPE - mean absolute percentage error.

````python
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true))
````

````python
mean_absolute_percentage_error(df['WAG_C_M'].iloc[history_length: history_length+len(date_list)].values,
                               df['forecast'].iloc[history_length: history_length+len(date_list)].values)
````

Построена модель SARIMA с отобранными оптимальными параметрами на основе начальных приближений: (5, 1, 5)x(0, 1, 0, 12season).

Построен прогноз заработной платы на год вперед, оценено качество прогноза с помощью средней абсолютной процентной ошибки MAPE. MAPE составляет 4.23%
